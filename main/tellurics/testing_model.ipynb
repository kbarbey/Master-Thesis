```python
model= np.loadtxt('model.dat')
data = hdulist = fits.open('test.fits')
y_data = hdulist[0].data
x_data = hdulist[0].header['CRVAL1'] + hdulist[0].header['CDELT1'] * np.arange(hdulist[0].header['NAXIS1'])

y_corrected = []
for i, y in enumerate(y_data):
    if np.isclose(model.T[1][i],0.0,atol=0.5):
        y_corrected.append(y)
    else:
        y_corrected.append(y / model.T[1][i])
```

```python
plt.figure(1)
plt.plot(x_data,y_data,'b')
plt.plot(x_data, y_corrected,'r',alpha=0.5)
```
